```python
# Set the date column as the DataFrame index
# Set the date column as the DataFrame index
# df_date_indexed = df.set_index("date")
# df_date_indexed.head()
df.set_index(df["date"], inplace = True)
df.head()
```

```python
# Drop the date column
df.pop("date")
df.head()
```

### Compare June and December data across all years 

```python
# from scipy import stats
```

```python
# Filter data for desired months
june_df = df.filter(like = '-06-', axis=0)
june_df
```

```python
dec_df = df.filter(like = '-12-', axis=0)
dec_df
```

```python
# Identify the average temperature for June
avg_june = round(june_df['tobs'].mean())
avg_june
```

```python
# Identify the average temperature for December
avg_dec = round(dec_df['tobs'].mean())
avg_dec
```

```python
# Create collections of temperature data
jun = june_df['tobs']
dec = dec_df['tobs']
```

```python
# Run paired t-test
t_value, p_value = stats.ttest_ind(jun, dec)
print(f"The t_value is {t_value}.")
print(f"The p_value is {p_value}.")
```
